# Calculating the confusion matrix
This notebook was used to process our output files from our BERT training sessions and analyze the results by calculating confusion matrixes and printing out the model performance measurements.

```python
import numpy as np, pandas as pd
from sklearn.metrics import confusion_matrix
```

```python
df_default = pd.read_json('20230414_160658_default.json')
df_emoji = pd.read_json('20230414_082501_emoji.json')
df_slang = pd.read_json('20230414_205935_slang.json')
df_slang_emoji = pd.read_json('20230426_002347_offensive.bak_results.json')
df_sample_data = pd.read_json('20230426_121152_eval_results.json')
```

```python
df_default.drop(['labels', 'predictions', 'lr','batch_size','epochs'], axis=1)
```

```python
Y_test = np.array(df_default.labels.tolist()[1])
y_pred = np.array(df_default.predictions.tolist()[1])
y_preds = np.argmax(y_pred, axis = 1).flatten()

from sklearn.metrics import ConfusionMatrixDisplay

labels = ["Offensive", "Not offensive"]
cm = confusion_matrix(Y_test, y_preds)

disp=ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
print('Confusion matrix for default setup')
disp.plot()
```
